```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
Names = []
Prices = []
Desc = []
Reviews = []

for i in range(1, 21):
    url = 'https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_8_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_8_0_na_na_na&as-pos=8&as-type=TRENDING&suggestionId=mobiles&requestId=7d114e70-3939-4503-a9db-4683e91a3b9b&as-searchtext=mobiles+5g&page=' + str(i)
    r = requests.get(url)
    print(r)

    
   
    soup = BeautifulSoup(r.text, 'lxml')
    box = soup.find('div', class_='_1YokD2 _3Mn1Gg')
    #find names
    names = box.find_all('div', class_ = '_4rR01T')
    #print(names)

    for i in names:
        n = i.text
        Names.append(n)
    print(len(Names))
    
    # find price
    prices = box.find_all('div', class_ = '_30jeq3 _1_WHN1')
    #print(prices)

    for i in prices:
        p = i.text
        Prices.append(p)
    print(len(Prices))
    
    # find description
    desc = box.find_all('ul', class_='_1xgFaf')
    #print(desc)

    for i in desc:
        d = i.text
        Desc.append(d)
    
    print(len(Desc))
    
    # find reviews
    reviews = box.find_all('div', class_ = '_3LWZlK')
    #print(reviews)

    for i in reviews:
        r = i.text
        Reviews.append(r)
    print(len(Reviews))
```

###  Scraping multiple pages


```python
df = pd.DataFrame({'Prod Name':Names,
                  'Prod Price':Prices,
                  'Prod Description':Desc,
                  'Prod Reviews': Reviews})
print(df)
```

```python
df.to_csv('devices.csv')
```
